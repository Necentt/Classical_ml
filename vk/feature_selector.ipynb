```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('trainee_train.csv')
train_df = train_df.iloc[:, 1:]
train_df.head()
```

```python
X = train_df.iloc[:, 1:]
y = train_df['im']
```

```python
corr_matrix = X.corr().abs()
```

```python
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.71)]
```

```python
X = X.drop(to_drop, axis=1)
```

```python
X.columns
```

```python
test_df = pd.read_csv('trainee_test_fish.csv')
test_df = test_df.iloc[:, 1:]
test_df = test_df.drop(to_drop, axis=1)
test_df.head()
```

```python
idx = features = X.columns.values
for df in [X, test_df]:
    df['sum'] = df[idx].sum(axis=1)
    df['min'] = df[idx].min(axis=1)
    df['max'] = df[idx].max(axis=1)
    df['mean'] = df[idx].mean(axis=1)
    df['std'] = df[idx].std(axis=1)
    df['skew'] = df[idx].skew(axis=1)
    df['kurt'] = df[idx].kurtosis(axis=1)
    df['med'] = df[idx].median(axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=42, shuffle=True, test_size=0.1)
```

```python
from sklearn.preprocessing import PolynomialFeatures

pf = PolynomialFeatures()
X_poly = pf.fit_transform(X_train)
```

```python
X_poly_val = pf.transform(X_val)
```

```python
X_val.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
```

only poly

```python
rf_clf = RandomForestClassifier()
rf_clf.fit(X_poly, y_train)
y_pred = rf_clf.predict(X_poly_val)
roc_auc_score(y_val, y_pred)
```

without poly

```python
rf_clf = RandomForestClassifier()
rf_clf.fit(X_train, y_train)
y_pred = rf_clf.predict(X_val)
roc_auc_score(y_val, y_pred)
```

poly + agg

```python
rf_clf = RandomForestClassifier()
rf_clf.fit(X_poly, y)
y_pred = rf_clf.predict(X_poly_val)
roc_auc_score(y_val, y_pred)
```
